# Challenge 1: Data Exploration

### Setup the Environment and Load the Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pl
import numpy as np

# Load the dataset (replace 'path_to_your_dataset.csv' with your actual file path)
df = pd.read_csv('synthetic_covid_impact_on_work.csv')

# Display the first few rows of the dataset
print(df.head())
```

#


###  Understand the Dataset Structure

```python
# Display the shape of the dataset (number of rows and columns)
print("Dataset shape:", df.shape)

# Display the columns in the dataset
print("Columns in dataset:", df.columns)
```

#

###  Check for Missing Values

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

#

### Generate Summary Statistics

```python
# Generate summary statistics for numerical features
summary_statistics = df.describe()
print("Summary Statistics:\n", summary_statistics)
```

#


#### Histogram: For understanding the distribution of a numerical feature like Work Hours.


```python
# Histogram of Increased Work Hours
plt.figure(figsize=(10, 6))
sns.histplot(df['Increased_Work_Hours'], bins=30, kde=True)
plt.title('Distribution of Increased Work Hours')
plt.xlabel('Work_From_Home')
plt.ylabel('Frequency')
plt.show()
```

#### Box Plot: To visualize outliers in features like Stress Levels.

```python
# Box plot of Stress Levels
plt.figure(figsize=(10, 6))
sns.boxplot(x='Stress_Level', data=df)
plt.title('Box Plot of Stress Levels')
plt.xlabel('Stress_Level')
plt.show()
```

# Challenge 2: Correlation Analysis

```python
print(df.head())
```

# 

### Calculate the Correlation Matrix

```python
print(df.apply(lambda col: col.unique()))
```

```python
import pandas as pd
import numpy as np

# Replace 'Low', 'High', etc., with NaN
df.replace({'Low': np.nan, 'High': np.nan}, inplace=True)

# Convert all columns to numeric, coercing errors
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
print(df.head())
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Display the correlation matrix
print("Correlation Matrix:\n", correlation_matrix)
```

#

#### Visualize the Correlation Matrix

#### A heatmap is an effective way to visualize the correlation matrix, allowing you to easily identify strong and weak correlations:

```python
# Set the size of the plot
plt.figure(figsize=(12, 8))

# Create a heatmap to visualize the correlation matrix
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Set the title and display the plot
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# 

### Analyze Key Correlations

Focus on interpreting the correlations relevant to the pandemic's impact on work patterns. For instance, consider the following:

Increased Work Hours vs. Stress Levels: A high positive correlation might suggest that as work hours increase, stress levels also rise.

Work from Home vs. Productivity Changes: A strong positive correlation could indicate that individuals who adapted to remote work saw improvements in productivity.

Work Hours vs. Job Satisfaction: Understanding how these factors interact can inform strategies for employee wellbeing.

```python
# Extract correlations of interest
target_features = ['Increased_Work_Hours', 'Stress_Level', 'Productivity_Change', 'Work_From_Home']
correlations_of_interest = correlation_matrix.loc[target_features, target_features]
print("Correlations of Interest:\n", correlations_of_interest)
```

```python
# Scatter plot for Increased Work Hours vs. Stress Levels
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Increased_Work_Hours', y='Stress_Level', data=df)
plt.title('Increased Work Hours vs. Stress Level')
plt.xlabel('Increased_Work_Hour')
plt.ylabel('Stress_Level')
plt.show()

# Scatter plot for Work from Home vs. Productivity Changes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Work_From_Home', y='Productivity_Change', data=df)
plt.title('Work_From_Home vs. Productivity_Change')
plt.xlabel('Work_From_Home')
plt.ylabel('Productivity_Change')
plt.show()
```
